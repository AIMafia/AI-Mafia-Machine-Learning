```python
import gensim
from gensim.models import word2vec
from gensim.models import KeyedVectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
word_vectors = KeyedVectors.load_word2vec_format("C:\\Users\\hp\\Downloads\\file1\\GoogleNews-vectors-negative300.bin",binary=True)
```

```python
def oddwords(words):
    words_to_check=[word_vectors[w] for w in words]
    average=np.mean(words_to_check,axis=0)
    odd_out=None
    min_similarity=1
    for w in words:
        sim=cosine_similarity([word_vectors[w]],[average])
        if sim<min_similarity:
            min_similarity=sim
            odd_out=w
    print("Similairy btw %s and avg vector is %.2f"%(w,sim))  
    return odd_out
```

```python
oddwords(["ugly","bad","beautiful"])
```

```python
def word_analogy(a,b,c,word_vectors):
    a,b,c=a.lower(),b.lower(),c.lower()
    max_similarity=-100
    d=None
    words = word_vectors.vocab.keys()
    
    wa,wb,wc = word_vectors[a],word_vectors[b],word_vectors[c]
    for w in words:
        if w in [a,b,c]:
            continue
        wv=word_vectors[w]
        sim=cosine_similarity([wb-wa],[wv-wc])
        if sim > max_similarity:
            max_similarity = sim
            d = w
            
    return d    
```

```python
word_analogy("man","woman","king",word_vectors)
```
