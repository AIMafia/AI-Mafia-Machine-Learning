```python
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(x,y,query,k=5):
    vals=[]
    for i in range(x.shape[0]):
        d=distance(query,x[i])
        vals.append((d,y[i]))
    vals=sorted(vals)
    vals=vals[:k]
    vals=np.array(vals)
    new_values=np.unique(vals[:,1],return_counts=True)
    index=new_values[1].argmax()
    pred=new_values[0][index]
    return pred
```

```python
m=test.shape[0]
for i in range(m):
    pred=knn(x,y,test[i])
    print(pred)
```
