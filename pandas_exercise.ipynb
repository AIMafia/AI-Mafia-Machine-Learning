```python
import datetime
d=[]
import calendar
def findDay(date):
    date=datetime.datetime.strptime(date,"%d-%m-%Y %H %M %S").weekday()
    return (calendar.day_name[date])
for i in range(0,len(df2["Date"])):
    d.append(findDay(df2["Date"][i].strftime("%d-%m-%Y %H %M %S")))
    
df2["WeekDay"]=d    
```
